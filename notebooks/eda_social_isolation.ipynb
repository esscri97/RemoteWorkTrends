## Pasos para explorar la relación:

1. Filtraremos los datos para centrarnos en empleados en modalidad remota (Work_Location == "Remote").

2. Analizaremos cómo varían los índices de satisfacción con el trabajo remoto dependiendo de los niveles de aislamiento social (Social_Isolation_Rating).

3. Visualizaremos los datos utilizando gráficos para identificar patrones.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset limpio
cleaned_path = '../data/processed/cleaned_data.csv'
df = pd.read_csv(cleaned_path)

# Crear un gráfico para comparar el aislamiento social según la modalidad de trabajo
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='Work_Location',
    y='Social_Isolation_Rating',
    order=['Remote', 'Hybrid', 'Onsite'],  # Ordenar modalidades
    palette="coolwarm"
)
# Guardar gráfico de aislamiento social según modalidad
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Work_Location', y='Social_Isolation_Rating', palette="coolwarm")
plt.title("Aislamiento Social según la Modalidad de Trabajo", fontsize=16)
plt.xlabel("Modalidad de Trabajo", fontsize=12)
plt.ylabel("Nivel de Aislamiento Social (1 - Bajo, 5 - Alto)", fontsize=12)
plt.savefig('../reports/social_isolation_analysis.png')  # Guardar en reports
plt.show()


# Estadísticas descriptivas para analizar tendencia de aislamiento por modalidad
isolation_stats = df.groupby('Work_Location')['Social_Isolation_Rating'].describe()
print("\nEstadísticas descriptivas del Aislamiento Social por Modalidad de Trabajo:")
print(isolation_stats)
```

## Conclusión

1. Distribución Similar entre Modalidades de Trabajo:

- Los niveles de aislamiento social (Social_Isolation_Rating) presentan una distribución muy similar en las tres modalidades: remoto, híbrido y presencial.

2. Implicaciones del Trabajo Remoto:

- Según este gráfico, trabajar de manera remota no se asocia directamente con niveles más altos de aislamiento social en comparación con trabajar de manera híbrida o presencial.

- Esto puede indicar que, aunque el trabajo remoto implica menor interacción física, no necesariamente incrementa la percepción de aislamiento social.
